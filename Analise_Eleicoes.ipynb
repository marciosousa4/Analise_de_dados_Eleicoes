```
# Importação das Bibliotecas
import pandas as pd
import matplotlib
```

```
# 2 - Criação do dataframe
df=pd.read_csv('votacao_partido_munzona_2016_BRASIL.csv', encoding = 'latin-1', sep = ';')
```

```
#Descrição das colunas do Dataframe
df.info()
```

```
df.sample(28)
```

```
# Separação dos dados por partido SG_Partido
df.groupby('SG_PARTIDO').size().sort_values()
```

```
#Representação Gráfica 
df.groupby('SG_PARTIDO').size().sort_values().tail(20).plot(kind='barh', figsize=(10,10))
```

```
# Criação de um novo Dataframe de comparação
compare = df[(df['SG_PARTIDO']=='PT')|(df['SG_PARTIDO']=='PSDB')]
```

```
#Dados dos partidos PT e PSDB
compare ['NM_PARTIDO'].value_counts()
```
